```python
import pandas as pd
import numpy as np
import re
import nltk
```

```python
df = pd.read_csv("IMDB_Top250Engmovies2_OMDB_Detailed.csv")
df.head()
```

```python
# converting to lowercase and removing numbers, punctuations, spaces, etc.
df['clean_plot'] = df['Plot'].str.lower()
df['clean_plot'] = df['clean_plot'].apply(lambda x: re.sub('[^a-zA-Z]', ' ', x))
df['clean_plot'] = df['clean_plot'].apply(lambda x: re.sub('\s+', ' ', x))
df['clean_plot']
```

```python
df['clean_plot'] = df['clean_plot'].apply(lambda x: nltk.word_tokenize(x))
df['clean_plot']
```

```python
# removing stopwords
stop_words = nltk.corpus.stopwords.words('english')
plot = []
for sentence in df['clean_plot']:
    temp = []
    for word in sentence:
        if word not in stop_words and len(word) >= 3:
            temp.append(word)
    plot.append(temp)
plot
```

```python
df['clean_plot'] = plot
```

```python
df['clean_plot']
```

```python
df.head(3)
```

```python
df['Genre'] = df['Genre'].apply(lambda x: x.split(','))
df['Actors'] = df['Actors'].apply(lambda x: x.split(',')[:4])
df['Director'] = df['Director'].apply(lambda x: x.split(','))
df['Production']=df['Production'].apply(lambda x: x.split(','))
```

```python
df['Actors'][0]
```

```python
def clean(sentence):
    temp = []
    for word in sentence:
        temp.append(word.lower().replace(' ', ''))
    return temp
```

```python
df['Genre'] = [clean(x) for x in df['Genre']]
df['Actors'] = [clean(x) for x in df['Actors']]
df['Director'] = [clean(x) for x in df['Director']]
df['Production'] = [clean(x) for x in df['Production']]
```

```python
df['Actors'][0]
```

```python
# combining all the columns data
columns = ['clean_plot', 'Genre', 'Actors', 'Director','Production']
tags = []
for i in range(len(df)):
    words = ''
    for col in columns:
        words += ' '.join(df[col][i]) + ' '
    tags.append(words)
tags
```

```python
df['clean_input'] = tags
df = df[['Title', 'clean_input']]
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
tfidf = TfidfVectorizer()
feat = tfidf.fit_transform(df['clean_input'])
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
df['clean_input'].apply(stem)
```

```python
df['clean_input']=df['clean_input'].apply(stem)
```

```python
def recommend(movie):
    movie_index = df[df['Title'] == movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(df.iloc[i[0]].Title)
```

```python
recommend('The Dark Knight Rises')
```
